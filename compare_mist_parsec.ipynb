## Read MIST isochrones

```python
ifn_mist = 'data/MIST_v1.2_vvcrit0.0_basic_isos/MIST_v1.2_feh_p0.00_afe_p0.0_vvcrit0.0_basic.txt'
t1 = read_mist_models.ISO(ifn_mist)
t1.hdr_list
print('version: ', t1.version)
print('abundances: ', t1.abun)
print('rotation: ', t1.rot)
print('ages: ', [round(x, 5) for x in t1.ages])
print('columns', t1.hdr_list)
```

## Read Parsec isochrones

```python
ifn_parsec = 'data/parsec/output123546912175.dat'
t2 = Table.read(ifn_parsec, format='ascii')
print('ages:',[round(x, 5) for x in sorted(set(t2['logAge']))])
display(t2[:100])
```

## Compare isochrones

```python
plt.close()
plt.rc('font', size=14)
plt.figure(1, figsize=(8, 6), tight_layout=True)

#Specify one or more ages
ages = [9.65]

#MIST
for age in ages: #age is in log scale
    i = t1.age_index(age) #returns the index for the desired age
    x = 10**t1.isos[i]['log_Teff']
    y = 10**t1.isos[i]['log_L']
    z = 10**t1.isos[i]['log_R']
    plt.scatter(x, y, s=z, color='b', lw=1, alpha=0.2,label='MIST')

#Parsec
#for age in list(set(t2['logAge']))[:2]:
for age in ages:
    m = abs(t2['logAge'] - age) < 0.02
    x = 10**t2[m]['logTe']
    y = 10**t2[m]['logL']
    z = t2[m]['Rpol']
    plt.scatter(x, y, s=z, color='r', lw=1, alpha=0.2,label='Parsec')    
    
plt.gca().invert_xaxis()
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$T_\\mathrm{eff}$ [K]')
plt.ylabel('$L$ [L$_\\odot$]')
plt.legend(loc=3)
plt.title(f'Isochrone at [Fe/H]={t1.abun["[Fe/H]"]}, age = {10**t1.ages[i]/1e9:4.2f} Ga')
plt.xlim(8e3, 2e3)
plt.show()

print(f'MIST:   Zinit = {t1.abun["Zinit"]}')
print(f'Parsec: Zinit = {t2["Zini"][0]}')

print(f'MIST:   log age = {t1.ages[i]}, age = {10**t1.ages[i]/1e9} Ga')
print(f'Parsec: log age = {t2[m]["logAge"][0]}, age = {10**t2[m]["logAge"][0]/1e9} Ga')
```
